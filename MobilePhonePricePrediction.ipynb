```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Mobile phone price.csv')
df.columns
```

```python
for i in range(len(df['Price ($)'])):
    a=""
    s=df['Price ($)'][i]
    for j in range(len(s)):
        if(s[j]!='$' and s[j]!=','):
            a=a+s[j]
        elif(s[j]==' '):
            break
    df['Price ($)'][i]=a
df['Price ($)']
```

```python
df['Price ($)']= df['Price ($)'].astype('int64')
df['Price ($)'].dtype
```

```python
len(df['Model'][0])
```

```python
X= df.iloc[:,0:7]

X['Brand'] = pd.factorize(X['Brand'])[0]
X['Model'] = pd.factorize(X['Model'])[0]
```

```python
print(X.columns)
```

```python
X['Storage'] = X['Storage'].str.replace('GB', '')
X['Storage'] = X['Storage'].astype('int64')
X['RAM'] = X['RAM'].str.replace('GB', '')
X['RAM'] = X['RAM'].astype('int64')
```

```python
X['Screen Size (inches)']=X['Screen Size (inches)'].astype(float)
print(X['Screen Size (inches)'].dtype)
```

```python
for i in range(len(X['Camera (MP)'])):
    a=""
    s=X['Camera (MP)'][i]
    for j in range(len(s)):
        if(s[j]!='+' and s[j]!='M'):
            a=a+s[j]
        else:
            break
    X['Camera (MP)'][i]=a
X['Camera (MP)']
```

```python
from sklearn.model_selection import train_test_split  
x_train, x_test, y_train, y_test= train_test_split(X, Y, test_size= 0.25, random_state=0)  
  
# feature Scaling  
from sklearn.preprocessing import StandardScaler    
st_x= StandardScaler()  
x_train= st_x.fit_transform(x_train)    
x_test= st_x.transform(x_test)
```

```python
from sklearn.tree import DecisionTreeRegressor 
  
regressor = DecisionTreeRegressor(random_state = 0) 
  
regressor.fit(x_train,y_train)
```

```python
y_pred_list = []

for row in x_test:
    
    row = row.reshape(1, -1)
    
    y_pred = regressor.predict(row)
    print(y_pred)
    
    y_pred_list.append(y_pred[0])


y_pred_array = np.array(y_pred_list)
```

```python
from sklearn.metrics import mean_absolute_error, r2_score


mae = mean_absolute_error(y_test, y_pred_array)

r2 = r2_score(y_test, y_pred_array)

accuracy = (1 - mae / y_test.mean()) * 100

print("Accuracy: {:.2f}%".format(accuracy))
print("R-squared score: {:.4f}".format(r2))
```
